```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

```python
def selu(x, lambdaa = 1.0507, alpha = 1.6732):
    return np.where(x<0, lambdaa*alpha*(np.exp(x)-1),x)
y = selu(x)
plot_graph(y,"SELU(X)")
```

```python
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
y = elu(x)
plot_graph(y,"ELU(X)")
```
